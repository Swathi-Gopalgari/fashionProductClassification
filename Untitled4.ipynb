```python
(x_train,y_train),(x_test,y_test)=tf.keras.datasets.fashion_mnist.load_data()
```

```python
class_labels=["T-shirt/top","Trowser","Pullover","Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankle boot"]
class_labels
```

```python
x_train =np.expand_dims(x_train,-1)
x_test=np.expand_dims(x_test,-1)
```

```python
x_train.ndim
```

```python
x_train=x_train/225
x_test=x_test/225
```

```python
from sklearn.model_selection import train_test_split
x_train,x_validation,y_train,y_validation=train_test_split(x_train,y_train,test_size=0.2,random_state=2020)
```

```python
x_train.shape,y_train.shape,x_validation.shape,y_validation.shape
```

```python
cnn_model=keras.models.Sequential([
    tf.keras.layers.Conv2D(filters=32, kernel_size=3, strides=(1,1),padding='valid',activation= 'relu',input_shape=[28,28,1]),
    tf.keras.layers.MaxPooling2D(pool_size=(2,2)),
    tf.keras.layers.Conv2D(filters=64,kernel_size=3, strides=(2,2),padding='same',activation= 'relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2,2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=128,activation='relu'),
    tf.keras.layers.Dropout(0,25),
    tf.keras.layers.Dense(units=256,activation='relu'),
    tf.keras.layers.Dropout(0,25),
    tf.keras.layers.Dense(units=128,activation='relu'),
    tf.keras.layers.Dense(units=10,activation='softmax'),

])
```

```python
y_train = tf.keras.utils.to_categorical(y_train, num_classes=10)
y_validation = tf.keras.utils.to_categorical(y_validation, num_classes=10)

cnn_model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
cnn_model.fit(x_train,y_train,epochs=28,batch_size=16,verbose=1,validation_data=(x_validation,y_validation))
```

```python
y_pred=cnn_model.predict(x_test)
```

```python
y_test = tf.keras.utils.to_categorical(y_test, num_classes=10)

# Now evaluate the model
cnn_model.evaluate(x_test, y_test)
```

```python
plt.figure(figsize=(16,30))
j=1
for i in np.random.randint(0,1000,60):
  plt.subplot(10,6,j)
  j+=1
  plt.imshow(x_test[i].reshape(28,28),cmap='Greys')
  # Get the index of the true and predicted labels
  true_label = np.argmax(y_test[i])
  pred_label = np.argmax(y_pred[i])
  plt.title('Actual= {} / {} \nPredicted= {} / {}'.format(class_labels[true_label], true_label, class_labels[pred_label], pred_label))
  plt.axis('off')
```

```python
from sklearn.metrics import confusion_matrix

plt.figure(figsize=(16,9))
y_pred_labels=[np.argmax(label) for label in y_pred]
# Convert y_test to multiclass by selecting the index of the highest probability
y_test_labels = np.argmax(y_test, axis=1)
cm=confusion_matrix(y_test_labels, y_pred_labels)

sns.heatmap(cm,annot=True, fmt='d',xticklabels=class_labels,yticklabels=class_labels)
```

```python
from sklearn.metrics import classification_report
# Convert y_test to multiclass by selecting the index of the highest probability
y_test_labels = np.argmax(y_test, axis=1)
cr=classification_report(y_test_labels, y_pred_labels,target_names=class_labels)
print(cr)
```
